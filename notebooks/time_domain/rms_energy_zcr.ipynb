```python
import librosa 
import librosa.display 
import IPython.display as ipd
import matplotlib.pyplot as plt
import numpy as np
```

```python
debussy_file = "audio/debussy.wav"
redhot_file = "audio/redhot.wav"
duke_file = "audio/duke.wav"
```

```python
debussy, _ = librosa.load("audio/debussy.wav")
redhot, _ = librosa.load("audio/redhot.wav")
duke, _ = librosa.load("audio/duke.wav")
```

```python
FRAME_LENGHT = 1024
HOP_LENGTH = 512

rms_debussy = librosa.feature.rms(y=debussy, frame_length=FRAME_LENGHT, hop_length=HOP_LENGTH)[0]
rms_redhot = librosa.feature.rms(y=redhot, frame_length=FRAME_LENGHT, hop_length=HOP_LENGTH)[0] 
rms_duke = librosa.feature.rms(y=duke, frame_length=FRAME_LENGHT, hop_length=HOP_LENGTH)[0]








# Plot the RMS values
```

```python
rms_debussy.shape
```

```python
#plot the rms values

frames = range(0, rms_debussy.size)
t = librosa.frames_to_time(frames, hop_length=512)

plt.figure(figsize=(15,17))

plt.subplot(3,1,1)
librosa.display.waveshow(debussy, alpha = 0.5)
plt.plot(t,rms_debussy,color = 'r')
plt.title("Debussy")
plt.ylim(-1,1)

plt.subplot(3,1,2)

librosa.display.waveshow(redhot, alpha = 0.5)
plt.plot(t, rms_redhot, color='r')
plt.title("Red Hot")
plt.ylim(-1,1)

plt.subplot(3,1,3)
librosa.display.waveshow(duke, alpha = 0.5)
plt.plot(t, rms_duke, color='r')
plt.title("Duke")
plt.ylim(-1,1)
```

```python
def rms(signal, frame_length=1024, hop_length=512):

    rms_energy = []
    for i in range(0,len(signal),hop_length):
        rms_energy.append(np.sqrt(np.mean(signal[i:i+frame_length]**2)))

    return np.array(rms_energy)



```

```python
rms1_debussy = rms(debussy, frame_length=FRAME_LENGHT, hop_length=HOP_LENGTH)
rms1_redhot = rms(redhot, frame_length=FRAME_LENGHT, hop_length=HOP_LENGTH) 
rms1_duke = rms(duke, frame_length=FRAME_LENGHT, hop_length=HOP_LENGTH)
```

```python
#plot the rms values

frames = range(0, rms_debussy.size)
t = librosa.frames_to_time(frames, hop_length=512)

plt.figure(figsize=(15,17))

plt.subplot(3,1,1)
librosa.display.waveshow(debussy, alpha = 0.5)
plt.plot(t,rms1_debussy,color = 'g')
plt.title("Debussy")
plt.ylim(-1,1)

plt.subplot(3,1,2)

librosa.display.waveshow(redhot, alpha = 0.5)
plt.plot(t, rms1_redhot, color='g')
plt.title("Red Hot")
plt.ylim(-1,1)

plt.subplot(3,1,3)
librosa.display.waveshow(duke, alpha = 0.5)
plt.plot(t, rms1_duke, color='g')
plt.title("Duke")
plt.ylim(-1,1)
```

```python
#zeroCrossingRate

zcr_debussy = librosa.feature.zero_crossing_rate(y=debussy, frame_length=FRAME_LENGHT, hop_length=HOP_LENGTH)[0]
zcr_redhot = librosa.feature.zero_crossing_rate(y=redhot, frame_length=FRAME_LENGHT, hop_length=HOP_LENGTH)[0]
zcr_duke = librosa.feature.zero_crossing_rate(y=duke, frame_length=FRAME_LENGHT, hop_length=HOP_LENGTH)[0]
```

```python
#plot the zcr values , normalized to 0-1

frames = range(0, zcr_debussy.size)
t = librosa.frames_to_time(frames, hop_length=512)

plt.figure(figsize=(15,17))
plt.plot(t, zcr_debussy, color='g')
plt.plot(t, zcr_redhot, color='r')
plt.plot(t, zcr_duke, color='b')
plt.ylim(0,1)
plt.legend(['Debussy', 'Red Hot', 'Duke'])
plt.show()
```

```python
# Non normalized zcr values

frames = range(0, zcr_debussy.size)
t = librosa.frames_to_time(frames, hop_length=512)

plt.figure(figsize=(15,17))
plt.plot(t, zcr_debussy* FRAME_LENGHT, color='g')
plt.plot(t, zcr_redhot* FRAME_LENGHT, color='r')
plt.plot(t, zcr_duke* FRAME_LENGHT, color='b')
plt.ylim(0,500)
plt.legend(['Debussy', 'Red Hot', 'Duke'])
plt.show()
```
