```python
#calculate the amplitude envelope of the debussy, redhot, and duke audio files
```

```python
debussy # all sample in 1 dimentional list only
```

```python
# Non overlapping frame

def amplitude_envelop(signal , frame_size=1024):
    amplitude_envelop = []

    for i in range(0,len(signal),frame_size):

        max_value_amplitude = max(signal[i:i+frame_size])
        amplitude_envelop.append(max_value_amplitude)

    return np.array(amplitude_envelop)
```

```python
ae_debussy = amplitude_envelop(debussy)
```

```python
len(ae_debussy)
```

```python
debussy.size/1024
```

```python
ae_debussy
```

```python
# overalapping frame

def amplitude_envelop_overlapping(signal , frame_size=1024, hop_size=512):
    amplitude_envelop = []

    for i in range(0,len(signal),hop_size):

        max_value_amplitude = max(signal[i:i+frame_size])
        amplitude_envelop.append(max_value_amplitude)

    return np.array(amplitude_envelop)
```

```python

ae_overlapping_debussy = amplitude_envelop_overlapping(debussy)
```

```python
len(ae_overlapping_debussy)
```

```python
ae_overlapping_debussy
```

```python
ae_overlapping_debussy = amplitude_envelop_overlapping(debussy)
ae_overlapping_redhot = amplitude_envelop_overlapping(redhot)
ae_overlapping_duke = amplitude_envelop_overlapping(duke)








#overlapping frame
```

```python
frames = range(0, ae_overlapping_debussy.size)
t = librosa.frames_to_time(frames, hop_length=512)

plt.figure(figsize=(15,17))

plt.subplot(3,1,1)
librosa.display.waveshow(debussy, alpha = 0.5)
plt.plot(t,ae_overlapping_debussy,color = 'r')
plt.title("Debussy")
plt.ylim(-1,1)

plt.subplot(3,1,2)
librosa.display.waveshow(redhot, alpha = 0.5)
plt.plot(t,ae_overlapping_redhot,color = 'r')
plt.title("Red Hot")
plt.ylim(-1,1)

plt.subplot(3,1,3)
librosa.display.waveshow(duke, alpha = 0.5)
plt.plot(t,ae_overlapping_duke,color = 'r')
plt.title("Duke")
plt.ylim(-1,1)

plt.tight_layout()
plt.show()
```
