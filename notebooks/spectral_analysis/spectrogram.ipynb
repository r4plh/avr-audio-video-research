```python
import numpy as np
import matplotlib.pyplot as plt
import os
import librosa
import IPython.display as ipd
```

```python
scale_file = "audio/scale.wav"
debussy_file = "audio/debussy.wav"
redhot_file = "audio/redhot.wav"
duke_file = "audio/duke.wav"

scale, sr = librosa.load(scale_file)
debussy , _ = librosa.load(debussy_file)
redhot, _ = librosa.load(redhot_file)
duke, _ = librosa.load(duke_file)





```

```python
 FRAME_SIZE = 2048
 HOP_LENGTH = 512
```

```python
S_scale = librosa.stft(scale, n_fft=FRAME_SIZE, hop_length=HOP_LENGTH)
```

```python
S_scale.shape # No of frequency bins x No of frames
```

```python
S_scale[0][0]
```

```python
Y_scale = np.abs(S_scale) # S(frame , frequency) matrix for all frames and all freq
```

```python
Y_scale[0][0] # magnitude of complex number 
```

```python
def plot_spectrogram(Y, sr, hop_length, y_axis="linear"):
    plt.figure(figsize=(25, 10))
    librosa.display.specshow(Y, 
                             sr=sr, 
                             hop_length=hop_length, 
                             x_axis="time", 
                             y_axis=y_axis)
    plt.colorbar(format="%+2.f")
```

```python
plot_spectrogram(Y_scale, sr, HOP_LENGTH)
```

```python
 Y_log_scale = librosa.power_to_db(Y_scale)
```

```python
plot_spectrogram(Y_log_scale, sr, HOP_LENGTH)
```

```python
plot_spectrogram(Y_log_scale, sr, HOP_LENGTH, y_axis = 'log')
# c4 , c#4 , d4 , d#4 , e4 , f4 , f#4 , g4 , g#4 , a4 , a#4 , b4
```

```python
S_debussy = librosa.stft(debussy, n_fft=FRAME_SIZE, hop_length=HOP_LENGTH)
S_redhot = librosa.stft(redhot, n_fft=FRAME_SIZE, hop_length=HOP_LENGTH)
S_duke = librosa.stft(duke, n_fft=FRAME_SIZE, hop_length=HOP_LENGTH)


Y_debussy = librosa.power_to_db(np.abs(S_debussy) ** 2)
Y_redhot = librosa.power_to_db(np.abs(S_redhot) ** 2)
Y_duke = librosa.power_to_db(np.abs(S_duke) ** 2)

plot_spectrogram(Y_debussy, sr, HOP_LENGTH, y_axis="log")
plot_spectrogram(Y_redhot, sr, HOP_LENGTH, y_axis="log")
plot_spectrogram(Y_duke, sr, HOP_LENGTH, y_axis="log")
```
