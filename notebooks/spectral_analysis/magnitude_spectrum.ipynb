```python
 def magnitude_spectrum(signal, title, sr, freq_ratio = 1):
    fft = np.fft.fft(signal)
    magnitude = np.abs(fft)
    
    plt.figure(figsize=(18,5))
    frequency = np.linspace(0,sr,len(magnitude))

    num_freq_bins = int(len(frequency) * freq_ratio)
    frequency = frequency[:num_freq_bins]
    magnitude = magnitude[:num_freq_bins]
    
    plt.plot(frequency, magnitude)
    plt.title(title)
    plt.xlabel("Frequency (Hz)")
    plt.ylabel("Magnitude")

    plt.show()   
```

```python
magnitude_spectrum(violin_c4, "Violin C4", sr)
```

```python
magnitude_spectrum(piano_c5, "Piano C5", sr)
```
